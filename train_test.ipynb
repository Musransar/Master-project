```python
import os
import shutil

source_dir = "TB_256"  # replace this with the path to your current directory
target_dir = "Dataset/TB"  # replace this with the path to your new directory

# create the target directory if it doesn't already exist
if not os.path.exists(target_dir):
    os.makedirs(target_dir)

# loop through all the files in the source directory
for file_name in os.listdir(source_dir):
    full_path = os.path.join(source_dir, file_name)
    # check if the file is a text file
    if not file_name.endswith(".txt"):
        # copy the file to the target directory
        shutil.copy(full_path, target_dir)
```

```python
import os
import random
import shutil

# Set the path to your data directory
data_dir = '/content/drive/MyDrive/tbx11k-simplified/TB_256'

# Set the path to your output directory
output_dirr = '/content/drive/MyDrive/tbx11k-simplified/Yolo_data'
output_dir = os.path.join(output_dirr, 'TB_256')
os.makedirs(output_dir, exist_ok=True)
# Create the train and test directories
train_dir = os.path.join(output_dir, 'train')
os.makedirs(train_dir, exist_ok=True)
test_dir = os.path.join(output_dir, 'test')
os.makedirs(test_dir, exist_ok=True)
```

```python
# Get a list of all the image files
image_files = [f for f in os.listdir(data_dir) if f.endswith('.png')]

# Randomly shuffle the image files
random.shuffle(image_files)

# Calculate the number of images for the train and test sets
num_train = int(len(image_files) * 0.8)
num_test = len(image_files) - num_train

# Copy the image and text files to the train and test directories
for i, file in enumerate(image_files):
    # Determine whether this file should go in the train or test set
    if i < num_train:
        dest_dir = train_dir
    else:
        dest_dir = test_dir

    # Copy the image file
    src = os.path.join(data_dir, file)
    dest = os.path.join(dest_dir, file)
    shutil.copy(src, dest)

    # Copy the corresponding text file
    text_file = file.replace('.png', '.txt')
    src = os.path.join(data_dir, text_file)
    dest = os.path.join(dest_dir, text_file)
    shutil.copy(src, dest)
```
